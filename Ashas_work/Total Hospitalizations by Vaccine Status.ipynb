```python
# Create a figure and axis object 
fig1, ax1 = plt.subplots(figsize=(10, 6))

# Create a bar chart for Unvaccinated rate
ax1.bar(df[df['Vaccine status'] == 'Unvaccinated']['Month'], df[df['Vaccine status'] == 'Unvaccinated']['Rate'], color='blue', label='Unvaccinated')

# Create a bar chart for Vaccinated rate
ax1.bar(df[df['Vaccine status'] == 'Vaccinated']['Month'], df[df['Vaccine status'] == 'Vaccinated']['Rate'], color='green', label='Vaccinated')

# Create a bar chart for Vaccinated + Booster rate
ax1.bar(df[df['Vaccine status'] == 'Vaccinated + Booster']['Month'], df[df['Vaccine status'] == 'Vaccinated + Booster']['Rate'], color='orange', label='Vaccinated + Booster')

# Set x-axis label and rotate
ax1.set_xlabel('Month')
ax1.tick_params(axis='x', labelrotation=90)

# Set y-axis label
ax1.set_ylabel('Rate of Hospitalizations')

# Set chart title
ax1.set_title('Total Hospitalizations by Vaccine Status')

# Rotate x-axis tick labels to be vertical
ax1.set_xticklabels(df['Month'], rotation=90)

# Add legend
ax1.legend()

# Display the chart
plt.show()

plt.savefig('demo.png', transparent=True)
```
